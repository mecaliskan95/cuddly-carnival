You work for a production company that specializes in nostalgic styles. You want to do some research on movies released in the 1990's. You'll delve into Netflix data and perform exploratory data analysis to better understand this awesome movie decade!

## The data

| Column | Description |

| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

# Display the shape and first few rows of the dataset
print(f"Dataset shape: {netflix_df.shape}")
netflix_df.head()
```

What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).

```python
# Data Filtering: Filter for movies released in the 1990s
netflix_90s_df = netflix_df[(netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] < 2000)]

# Statistical Analysis: Calculate statistics for movie durations
mean_duration = netflix_90s_df['duration'].mean()
median_duration = netflix_90s_df['duration'].median()
mode_duration = netflix_90s_df['duration'].mode()[0]

# Print duration statistics
print(f"Mean Duration: {mean_duration}")
print(f"Median Duration: {median_duration}")
print(f"Mode Duration: {mode_duration}")

# Data Visualization: Create a histogram of movie durations
plt.figure(figsize=(10, 6))
plt.hist(netflix_90s_df['duration'], bins=40, edgecolor='black')
plt.title('Distribution of Movie Durations (1990s)', fontsize=14)
plt.xlabel('Duration (minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.axvline(mode_duration, color='r', linestyle='dashed', linewidth=1, label='Mode Duration')
plt.legend()
plt.show()
```

A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

```python
# Counting Short Action Movies: Count short action movies (less than 90 minutes)
short_movie_count = 0  # Initialize counter
for label, row in netflix_90s_df.iterrows():
    if row['genre'] == "Action" and row['duration'] < 90:
        short_movie_count += 1
        
# Print the count of short action movies
print(f"Number of short action movies from the 1990s: {short_movie_count}")
```

```python
# Conclusion: Print final results
print(f"The most frequent movie duration in the 1990s is approximately {mode_duration} minutes.")
print(f"A total of {short_movie_count} short action movies were released during this decade.")
```
