## Introduction

Hurricanes are major natural disasters that cause billions of dollars in property damage and significant loss of life in the United States each year. This study aims to reveal seasonal, annual, and geographical trends using data visualization techniques while examining hurricanes that occurred in the United States from 1950 to 2022. The findings aim to make significant contributions to the development of disaster management and emergency response strategies.

### Data Set

The dataset used for our analysis is based on the National Weather Service Storm Prediction Center Severe Weather Maps, Graphics, and Data Page provided by the U.S. National Oceanic and Atmospheric Administration (NOAA) and the file named ['tornados'](https://github.com/rfordatascience/tidytuesday/blob/master/data/2023/2023-05-16/readme.md) consisting of 68,693 records. This dataset includes 27 different features of hurricanes, such as date, intensity, track, and damage caused. The main features used in the study are explained below:

*   'yr' - integer - Ordinal - Year of hurricane occurrence, ranging from 1950 to 2022.
*   'st' - integer - Categorical - Two-letter postal abbreviation of the state (e.g., NY = New York, NJ = New Jersey).
*   'fat' - integer - Quantitative - Number of fatalities caused by the hurricane.

### Code

The Python programming language was used for data processing, and data pre-processing, analysis, and visualization operations were performed in the Google Colab environment with libraries such as pandas, numpy, and plotly. Missing and inconsistent data in the dataset were examined and cleaned if necessary, and made suitable for analysis.

## Data Reading

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression

# Load the dataset from the URL
tornados_url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2023/2023-05-16/tornados.csv'
df = pd.read_csv(tornados_url)
```

## Data Pre-processing and Analysis

```python
# Basic statistics
print(df.head())
print(df.info())
print(df.describe())

# Missing data check
print(df.isnull().sum())
```

## Question 1

### Question 1. How has the number of annual tornadoes in the US changed from 1950 to 2022, and in which years were the highest and lowest numbers of tornadoes observed during this period?

```python
# Group by year and calculate the number of tornadoes.
grouped = df.groupby('yr').size().reset_index(name='tornado_count')

# Find the years with the maximum and minimum number of tornadoes.
max_tornado = grouped.loc[grouped['tornado_count'].idxmax()]
min_tornado = grouped.loc[grouped['tornado_count'].idxmin()]

# Create a model for the trend.
X = grouped['yr'].values.reshape(-1, 1)
y = grouped['tornado_count'].values
model = LinearRegression().fit(X, y)
trendline = model.predict(X).round().astype(int)

# Create a dynamic line graph.
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=grouped['yr'],
    y=grouped['tornado_count'],
    mode='lines',
    line=dict(color='red', width=2),
    name='Tornado Count<br>Over Years',
    hovertemplate='<b>Year:</b> %{x}<br><b>Tornado Count:</b> %{y}<extra></extra>'
))

fig.add_trace(go.Scatter(
    x=grouped['yr'],
    y=trendline,
    mode='lines',
    line=dict(color='yellow', dash='dash'),
    name='Trend Line',
    hoverinfo='none'
))

# Label the years with the most and least tornadoes.
fig.add_trace(go.Scatter(
    x=[max_tornado['yr'], min_tornado['yr']],
    y=[max_tornado['tornado_count'], min_tornado['tornado_count']],
    mode='markers+text',
    marker=dict(color='red', size=6, symbol='circle', line=dict(color='blue', width=2)),
    text=['Highest Tornado Count', 'Lowest Tornado Count'],
    textposition=['middle left', 'bottom right'],
    textfont=dict(color='white'),
    hoverinfo='text',
    hovertext=[
        f"<b>Year::</b> {max_tornado['yr']}<br><b>Tornado Count:</b> {int(max_tornado['tornado_count'])}",
        f"<b>Year:</b> {min_tornado['yr']}<br><b>Tornado Count:</b> {int(min_tornado['tornado_count'])}"
    ],
    showlegend=True,
    name='Special Points'
))

# Update the layout of the graph (title, axes labels, range, template, legend, font).
fig.update_layout(
    title="Annual Tornado Count in the US (1950-2022): Years with Highest and Lowest Counts & Trend Analysis",
    xaxis_title='Year',
    yaxis_title='Tornado Count',
    xaxis=dict(
        range=[1949, 2023],
        tickmode='auto',
        tickformat="d",
        showgrid=False
    ),
    yaxis=dict(
        range=[0, 1920],
        tickmode='auto',
        tickformat="d",
        showgrid=False
    ),
    template='plotly_dark',
    legend=dict(
        x=0.98, y=0.05,
        xanchor='right', yanchor='bottom',
        bgcolor='rgba(0,0,0,0)'
    ),
    autosize=True,
    dragmode=False,
    font=dict(family="Arial", size=12, color="white")
)

# Disable zoom and pan features.
fig.update_geos(
    visible=False
)

# Show the figure with disabled display mode bar and double click.
fig.show(config=dict(displayModeBar=False, doubleClick=False))
```

### Figure 1: Line Graph Illustrating the Annual Number of Tornadoes Recorded in the US from 1950 to 2022. Changes in tornado activity provide critical information for disaster management and policy development, contributing to the creation of more effective decision-making mechanisms.

### Findings:
* The yellow dashed trend line indicates a general increasing trend in the number of tornadoes observed in the US from 1950 to 2022.
* Significant fluctuations in tornado counts have been observed throughout the analyzed period. The relatively low increase in the number of tornadoes from the 1950s to the 1960s may indicate that meteorological conditions were more stable during those years. However, the sudden increases and decreases starting from the early 2000s show the impact of climate change during this period and that tornado activity is more variable. This is an important factor to consider in disaster preparedness planning.
* The year 2004 stands out as the most intense year in the period under review, with 1817 tornadoes. The lowest number of tornadoes was recorded in 1950, with only 201 events. These are critical points showing significant changes in annual tornado activity.

### Visual Channels Used:
* Data Mark: A line graph was used to show the change in the number of tornadoes by year.
* X-Axis: Years, from 1950 to 2022 (Ordinal)
* Y-Axis: Number of Tornadoes (Quantitative)
* Interaction: When hovering over the graph, details of the number of tornadoes for a specific year can be seen.
* Label: Important data points, such as "Highest Number of Tornadoes" and "Lowest Number of Tornadoes", have been clearly marked to attract the user's attention and convey information quickly.
* Symbol: Special data points (records with the highest and lowest number of tornadoes) are further highlighted with red dots with blue borders.
* Color and Line Style: The main data series is indicated in red, and the trend line is shown with a yellow dashed line.


## Question 2

### Question 2. How are the total number of deaths caused by tornadoes distributed across different states in the US from 1950 to 2022?

```python
# # Create a dictionary mapping state abbreviations to full names
state_names = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas', 'CA': 'California',
    'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware', 'FL': 'Florida', 'GA': 'Georgia',
    'HI': 'Hawaii', 'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa', 'KS': 'Kansas',
    'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland', 'MA': 'Massachusetts',
    'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi', 'MO': 'Missouri', 'MT': 'Montana',
    'NE': 'Nebraska', 'NV': 'Nevada', 'NH': 'New Hampshire', 'NJ': 'New Jersey', 'NM': 'New Mexico',
    'NY': 'New York', 'NC': 'North Carolina', 'ND': 'North Dakota', 'OH': 'Ohio', 'OK': 'Oklahoma',
    'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina',
    'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah', 'VT': 'Vermont',
    'VA': 'Virginia', 'WA': 'Washington', 'WV': 'West Virginia', 'WI': 'Wisconsin', 'WY': 'Wyoming',
    'DC': 'District of Columbia', 'PR': 'Puerto Rico', 'VI': 'Virgin Islands'
}

# Calculate fatalities per state
state_fatalities = df.groupby('st')['fat'].sum().reset_index()

# Add state names as a new column
state_fatalities['state_name'] = state_fatalities['st'].map(state_names)

# Create a choropleth map with Plotly
fig = px.choropleth(state_fatalities,
                    locations='st',
                    locationmode='USA-states',
                    color='fat',
                    hover_name='state_name',
                    hover_data={'st': False, 'fat': True, 'state_name': False},
                    color_continuous_scale='Reds',
                    scope='usa',
                    title="Total Tornado Fatalities by State in the US (1950-2022)",
                    labels={'fat': 'Total<br>Fatalities', 'st': 'State'},
                    template='plotly_dark')

# Adjust hover data format
fig.update_traces(
    hovertemplate='<b>State:</b> %{hovertext}<br><b>Fatalities:</b> %{customdata[1]}<extra></extra>',
    customdata=state_fatalities[['state_name', 'fat']].values,
    hovertext=state_fatalities['state_name']
)

# Update graph settings
fig.update_layout(
    title="Total Tornado Fatalities by State in the US (1950-2022)",
    geo=dict(
        projection_type='albers usa',
        showframe=False,
        showcoastlines=False,
        projection_scale=1,
    ),
    margin=dict(l=0, r=0, t=75, b=25),
    autosize=True,
    dragmode=False,
    font=dict(family="Arial", size=12, color="white")
)

# Disable zoom and pan features
fig.update_geos(
    visible=False
)

fig.show(config=dict(displayModeBar=False, doubleClick=False))
```

### Figure 2. Choropleth Map Illustrating Total Tornado Fatalities by State in the US from 1950 to 2022. The color scale represents the number of fatalities in each state, with darker shades indicating higher death tolls. This visual emphasizes the importance of intensifying tornado preparedness and response efforts, especially in high-risk states.

### Findings:
* The choropleth map reveals high fatality rates in the central US region known as "Tornado Alley," encompassing states like Alabama, Texas, Mississippi, Oklahoma, Tennessee, Arkansas, and Missouri [1]. These states, marked with dark red on the map, are highlighted as areas where tornadoes occur frequently and result in high casualties. Notably, Alabama stands out with the highest number of fatalities at 668.
* While regions like the Midwest and Southern states (e.g., Missouri, Tennessee, Kentucky) have relatively high fatality numbers, Western and Northeastern states have lower fatality rates. This is depicted on the map with lighter color shades, revealing regional differences in tornado preparedness and prevention measures.
* The findings indicate the need to strengthen tornado management and emergency response strategies, particularly in high-risk areas. This necessitates more effective measures and preparations to minimize casualties in regions like "Tornado Alley" where tornadoes are frequent.

### Visual Channels Used:
* Data Mark: A choropleth map was used to visualize the number of fatalities by state.
* X-Axis: Latitude (Quantitative - Labels not shown in the visual)
* Y-Axis: Longitude (Quantitative - Labels not shown in the visual)
Line: State borders are clearly drawn so users can easily distinguish the differences between regions.
* Interaction: When hovering over the map, the name of a specific state and the number of fatalities are displayed.
* Color: The number of fatalities is represented by color intensity; high numbers are shown in dark red, and low numbers are shown in lighter shades (Quantitative).




## Question 3

### Question 3. How have the total number of tornadoes by state changed over time in the US from 1950 to 2022, and which states have experienced the most or fewest tornadoes during this period?

```python
# Calculate tornado counts by state and year, replace abbreviations with full names
state_year_counts = df.groupby(['yr', 'st']).size().reset_index(name='count')
state_year_counts['st'] = state_year_counts['st'].apply(lambda x: state_names[x])

# Create a pivot table to calculate cumulative tornado counts
state_year_pivot = state_year_counts.pivot_table(index='st', columns='yr', values='count', fill_value=0)
state_year_cumsum = state_year_pivot.cumsum(axis=1).reset_index().melt(id_vars=['st'], var_name='yr', value_name='count')

# Determine ranking based on the last year and sort DataFrame accordingly
final_year_ranking = state_year_cumsum[state_year_cumsum['yr'] == state_year_cumsum['yr'].max()].groupby('st')['count'].sum().sort_values(ascending=False).index.tolist()
state_year_cumsum['fixed_rank'] = state_year_cumsum['st'].map({name: i for i, name in enumerate(final_year_ranking)})
state_year_cumsum.sort_values(['yr', 'fixed_rank'], inplace=True)

# Calculate maximum cumulative count and set x-axis range
max_cumulative_count = state_year_cumsum['count'].max()
range_x = [0, max_cumulative_count * 1.1]

# Calculate annual total tornado count and rank by state
state_year_total = state_year_cumsum.groupby(['yr', 'st'])['count'].sum().reset_index()
state_year_total['rank'] = state_year_total.groupby('yr')['count'].rank(ascending=False, method='first')

# Merge data to enable sorting states by tornado count
merged = state_year_cumsum.merge(state_year_total[['yr', 'st', 'rank']], on=['yr', 'st'])
merged = merged.sort_values(['yr', 'rank'])
merged['yr'] = merged['yr'].astype(int)

# Keep state ranking consistent with the final year
final_year_ranking.reverse()
merged['fixed_rank'] = merged['st'].map({name: i for i, name in enumerate(final_year_ranking)})
merged.sort_values(['yr', 'fixed_rank'], inplace=True)

# Create the animated bar chart
fig = px.bar(merged,
             x='count',
             y='st',
             animation_frame='yr',
             orientation='h',
             title='Total Tornado Count by State (1950-2022)',
             labels={'count': 'Total Tornado Count', 'st': 'States', 'yr': 'Year'},
             hover_data={'count': True, 'yr': True, 'st': False},
             range_x=[0, max_cumulative_count * 1.1],
             color_discrete_sequence=['#FFD700'])

# Fix y-axis order (highest to lowest)
fig.update_yaxes(categoryorder='array', categoryarray=final_year_ranking)

# Update layout (buttons, slider, height, margins, font)
fig.update_layout(
    template='plotly_dark',
    updatemenus=[{
        'buttons': [
            {
                'args': [None, {
                    'frame': {'duration': 1600, 'redraw': True},
                    'fromcurrent': True,
                    'transition': {'duration': 500, 'easing': 'linear'}
                }],
                'label': 'Play',
                'method': 'animate'
            },
            {
                'args': [[None], {
                    'frame': {'duration': 0, 'redraw': True},
                    'mode': 'immediate',
                    'transition': {'duration': 0}
                }],
                'label': 'Pause',
                'method': 'animate'
            }
        ],
        'direction': 'left',
        'pad': {'r': 10, 't': 87},
        'showactive': False,
        'type': 'buttons',
        'x': 0.025,
        'xanchor': 'right',
        'y': 0.05,
        'yanchor': 'top'
    }],
    sliders=[{
        'currentvalue': {
            'prefix': 'Year: ',
            'visible': True,
            'xanchor': 'right'
        },
        'steps': [{'args': [[str(year)], {'frame': {'duration': 300, 'redraw': True}, 'mode': 'immediate', 'transition': {'duration': 300}}],
                   'label': str(year), 'method': 'animate'} for year in merged['yr'].unique()],
        'x': 0.025,
        'len': 0.9,
        'pad': {'b': 10, 't': 30}
    }],
    height=1100,
    margin=dict(l=170, r=50, t=80, b=20),
    font=dict(family="Arial", size=12, color="white")
)

# Customize x-axis ticks
tickvals = [i for i in range(0, int(max_cumulative_count) + 100, 1000)]
ticktext = [f'{i:,}'.replace(',', '.') for i in tickvals]
fig.update_xaxes(tickvals=tickvals, ticktext=ticktext)
#fig.update_yaxes(autorange=True)

# Increase bar thickness and adjust colors
fig.update_traces(marker=dict(line=dict(width=1), opacity=0.7),
                  selector=dict(type='bar'))

fig.show()
```

### Figure 3. This animated horizontal bar chart shows the total number of tornadoes and annual changes in each state of the US from 1950 to 2022. Each bar represents the total number of tornadoes a state has encountered, and the animation highlights the change over time. This visualization helps us understand the geographical distribution of tornado activity and trends over time. The Y-axis is sorted from the state with the highest total number of tornadoes in 2022 to the lowest. Thus, changes and jumps over the years become observable.

### Findings:
* One of the areas where tornadoes are most intense in the United States is Tornado Alley, which includes states such as Texas, Oklahoma, Kansas, and Nebraska. Due to the regional climate conditions being conducive to tornado formation, this region has long been the focus of tornado activity [1]. According to visual data, Texas, Oklahoma, and Kansas have consistently stood out with high numbers of tornadoes from 1950 to 2022.
